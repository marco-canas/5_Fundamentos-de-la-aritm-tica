```python
import pandas as pd
import numpy as np

# Cargar la lista de estudiantes desde el archivo CSV
path = 'C:/Users/marco/Documentos/docencia/groups_list/fundamentos_aritmetica.csv'
estudiantes = pd.read_csv(path)
estudiantes['Parcial'] = estudiantes['Parcial'].astype(float)  # Asegurar que la columna 'Parcial' es float

# Convertir la columna 'Nombre' a una lista de Python
lista_estudiantes = estudiantes['Nombre'].tolist()

# Función para llamar a un estudiante aleatoriamente y registrar su nota de seguimiento
def asistencia():
    global lista_estudiantes
    if lista_estudiantes:
        estudiante_aleatorio = np.random.choice(lista_estudiantes)
        print(f"Llamando a {estudiante_aleatorio}...")
        asistencia = input(f"¿Asistió {estudiante_aleatorio}? (s/n): ").strip().lower()
        
        # Validación de entrada
        if asistencia == 's':
            # Obtener el índice del estudiante en el DataFrame
            indice = estudiantes[estudiantes['Nombre'] == estudiante_aleatorio].index[0]
            # Incrementar su valor en la columna 'Parcial'
            estudiantes.at[indice, 'Parcial'] += 0.3
            print(f"Actualizado: {estudiante_aleatorio} ahora tiene {estudiantes.at[indice, 'Parcial']} en la columna 'Parcial'.")
        
        elif asistencia != 'n':
            print("Entrada inválida. Por favor, introduce 's' para sí o 'n' para no.")
            return  # Volver a preguntar para el mismo estudiante si la entrada es inválida
        
        # Remover al estudiante de la lista para no llamarlo de nuevo
        lista_estudiantes.remove(estudiante_aleatorio)
    else:
        print("Se han llamado a todos los estudiantes del grupo completo.")

# Función para guardar cambios al final de la sesión
def guardar_cambios():
    estudiantes.to_csv(path, index=False)
    print("Cambios guardados en el archivo CSV.")

# Ejemplo de uso
while lista_estudiantes:
    asistencia()

# Guardar los cambios una vez que todos los estudiantes hayan sido llamados
guardar_cambios()

```

El enfoque STEAMS en educación preescolar, primaria y secundaria es una extensión del enfoque STEM (Ciencia, Tecnología, Ingeniería y Matemáticas) y su variante STEAM, que incorpora el Arte. 

El acrónimo STEAMS incluye:



1. **Science (Ciencia):** Fomenta la curiosidad por el mundo natural y el desarrollo de habilidades científicas, como la observación, experimentación y análisis.



2. **Technology (Tecnología):** Integra herramientas tecnológicas para potenciar el aprendizaje y la resolución de problemas, promoviendo la alfabetización digital desde edades tempranas.



3. **Engineering (Ingeniería):** Introduce el pensamiento lógico y la resolución de problemas estructurados, fomentando habilidades como el diseño y la construcción de soluciones.



4. **Arts (Arte):** Añade creatividad y pensamiento crítico, permitiendo a los estudiantes expresarse y explorar el aprendizaje a través de formas artísticas (plásticas, música, teatro).



5. **Mathematics (Matemáticas):** Refuerza las habilidades numéricas y el razonamiento lógico, fundamentales en la resolución de problemas y comprensión de conceptos abstractos.



6. **Social Sciences (Ciencias Sociales):** Incorpora la comprensión del entorno social, político, cultural y económico, promoviendo el desarrollo de la conciencia crítica sobre el mundo y su historia, además de habilidades de empatía y ciudadanía.



El enfoque STEAMS pretende desarrollar un aprendizaje **interdisciplinario** que integre estas áreas para fomentar un pensamiento más holístico y práctico. 

Se busca que los estudiantes no solo aprendan conocimientos técnicos, sino también cómo aplicarlos creativa y críticamente en situaciones reales, promoviendo habilidades como el trabajo en equipo, la comunicación y la innovación. 

En educación preescolar, primaria, secundaria y superior, se adapta a las diferentes etapas de desarrollo cognitivo y emocional de los estudiantes, promoviendo el aprendizaje activo y contextualizado.

## Hipótesis de los estudiantes Luis Fernando y Cristián  

¿Será que los números naturales terminados en 1 siguen el patrón Primo-Primo- No primo.?  



```python
def es_primo(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
```

```python
import numpy as np 

numeros_que_terminan_en_cifra_1 = np.arange(11, 10_000, 10)
numeros_que_terminan_en_cifra_1 
```

```python
[es_primo(n) for n in numeros_que_terminan_en_cifra_1] 
```

## Conclusión 
Se descarta la hipótesis de que los números naturales terminados en 1 siguen el pantron P-P-NP

Esta es una colección de 10 ejercicios de conjeturas sobre números enteros que pueden probar y validar usando la función `es_primo(n)`. Cada conjetura les permitirá explorar patrones y características de los números primos y otros tipos de números enteros.



1. **Conjetura de Números Gemelos**  
   Investigar si existen números primos gemelos para los primeros \( N \) números naturales, donde un par de números primos gemelos son primos consecutivos que tienen una diferencia de 2 (como 11 y 13).  
   **Ejemplo**: Validar que \( 17 \) y \( 19 \) son primos gemelos.

2. **Conjetura de Números Primoriales**  
   Explorar si todos los primos de la forma \( p_n! + 1 \) (factorial de un número primo más 1) son números primos.  
   **Ejemplo**: Probar con $p = 5$: \( 5! + 1 = 121 \).



3. **Conjetura de los Números de Mersenne**  
   Comprobar si los números de la forma \( 2^p - 1 \) (donde \( p \) es un número primo) son siempre primos. Estos números se conocen como números de Mersenne.  
   **Ejemplo**: Verificar si \( 2^3 - 1 = 7 \) es primo.



4. **Conjetura de Primos de la Forma \( n^2 + n + 41 \)**  
   Evaluar si la expresión \( n^2 + n + 41 \) produce números primos para \( n \) entre 0 y 39.  
   **Ejemplo**: Comprobar si \( 0^2 + 0 + 41 = 41 \) y \( 1^2 + 1 + 41 = 43 \) son primos.



5. **Conjetura de Primos de Fermat**  
   Verificar si los números de la forma \( 2^{2^n} + 1 \) son primos para $n = 0, 1, 2, 3, \ldots$.  
   **Ejemplo**: Comprobar si \( 2^{2^2} + 1 = 17 \) es primo.



6. **Conjetura de Números Semi-Primos**  
   Probar si un número compuesto se puede expresar siempre como el producto de dos números primos.  
   **Ejemplo**: Comprobar si $15 = 3 \times 5$.



7. **Conjetura de Primos de Sophie Germain**  
   Investigar si, dado un número primo \( p \), el número \( 2p + 1 \) también es primo. Estos números \( p \) se llaman primos de Sophie Germain.  
   **Ejemplo**: Comprobar si, siendo \( p = 5 \), \( 2 \times 5 + 1 = 11 \) es primo.



8. **Conjetura sobre Suma de Dos Primos**  
   Explorar si todo número par mayor que 2 puede expresarse como la suma de dos números primos (Conjetura de Goldbach).  
   **Ejemplo**: Comprobar si $10 = 3 + 7$ y si $18 = 5 + 13$.



9. **Conjetura de Números Triangulares y Primos**  
   Investigar si el número triangular \( T_n = \frac{n(n+1)}{2} \) es primo para algunos valores de $n$.  
   **Ejemplo**: Comprobar si \( T_2 = 3 \) y \( T_3 = 6 \) son primos.



10. **Conjetura de Números Cuadrados y Primos**  
    Comprobar si existe algún número primo que también sea un número cuadrado perfecto.  
    **Ejemplo**: Probar si \( 9 = 3^2 \) y es primo.

Cada uno de estos ejercicios fomenta la investigación y el razonamiento matemático en torno a las propiedades de los números enteros, permitiendo a tus estudiantes poner a prueba sus habilidades en la programación y en el análisis de patrones numéricos.

### **Actividad de Clase: Expresar Ideas, Dudas y Emociones de Manera Respetuosa y Asertiva**



**Objetivo:**  
Desarrollar habilidades de comunicación efectiva en los estudiantes para que puedan expresar sus ideas, dudas y emociones de forma respetuosa y asertiva.



**Duración:**  
15-20 minutos

---



### **1. Introducción (5 minutos)**  
- **Tema:** Introducir a los estudiantes al concepto de la comunicación asertiva, explicando cómo expresar pensamientos y emociones de manera clara y respetuosa sin agredir o ser pasivo.
- **Dinámica:** Mostrar dos ejemplos contrastantes: uno de comunicación pasiva/agresiva y otro de comunicación asertiva. Preguntar a los estudiantes cuál creen que es más efectivo y por qué.



**Materiales:**  
- Pizarra o diapositivas con ejemplos de frases pasivas, agresivas y asertivas.
  
**Ejemplo:**  
- **Pasivo:** “Bueno… no sé si es una buena idea, pero… tal vez podrías intentarlo.”
- **Agresivo:** “¡Tu idea es terrible, no va a funcionar!”
- **Asertivo:** “Creo que podríamos mejorar esa idea si consideramos otro enfoque.”

---



### **2. Ejercicio en Parejas (10 minutos)**  
- **Instrucciones:** Dividir a los estudiantes en parejas. Darles diferentes situaciones cotidianas en las que deben comunicarse (pedir ayuda, expresar desacuerdo, compartir una idea, etc.). Cada pareja debe crear una breve conversación donde practiquen la comunicación asertiva.
  
**Situaciones propuestas:**
  - Expresar desacuerdo con un compañero sobre una tarea de grupo.
  - Pedirle ayuda al profesor sobre un concepto difícil.
  - Expresar frustración por un malentendido en clase.

- **Reglas:** Cada estudiante debe asegurarse de usar lenguaje respetuoso y asertivo para expresar su posición y emociones. Al final, cada pareja debe compartir con la clase un ejemplo de lo que discutieron.

---



### **3. Retroalimentación y Discusión (5 minutos)**  
- **Discusión:** Invitar a algunos estudiantes a compartir sus experiencias de la actividad. 
- **Preguntas:** ¿Cómo se sintieron al intentar ser asertivos? ¿Les resultó difícil? ¿Cómo fue recibido el mensaje?
  
**Consejo:** Resaltar la importancia de practicar la asertividad como una herramienta para mejorar la comunicación interpersonal y reducir conflictos.

---



### **Cierre**  
- Resumir las características de la comunicación asertiva: uso de "yo" en lugar de "tú" para evitar culpabilizar, tono de voz calmado, y respeto hacia el interlocutor.
- **Tarea opcional:** Pedirles que practiquen la comunicación asertiva en sus conversaciones durante la semana y escriban una reflexión breve sobre cómo les fue.



**Material adicional:**  
- Video corto o recurso digital sobre la importancia de la comunicación asertiva.
