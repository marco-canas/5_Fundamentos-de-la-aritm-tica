````python
# algoritmo para llamar a evaluación oral individual
import pandas as pd
import numpy as np

# Cargar la lista de estudiantes desde la URL
path = '/home/marco/Documentos/docencia/groups_list/calculo_vectorial.csv'
estudiantes = pd.read_csv(path)
estudiantes['Parcial'] = estudiantes['Parcial'].astype(float) 

# Convertir la columna 'Nombre' a una lista de Python
lista_estudiantes = estudiantes['Nombre'].tolist()

# Función para llamar a un estudiante aleatoriamente y registrar su nota de seguimiento
def eval_oral():
    if lista_estudiantes:
        estudiante_aleatorio = np.random.choice(lista_estudiantes)
        print(f"Llamando a {estudiante_aleatorio}...")
        asistencia = input(f"¿Se expresó bien y correco {estudiante_aleatorio}? (s/n): ").strip().lower()
        
        if asistencia == 's':
            # Obtener el índice del estudiante en el DataFrame
            indice = estudiantes[estudiantes['Nombre'] == estudiante_aleatorio].index[0]
            # Incrementar su valor en la columna 'Parcial'
            estudiantes.at[indice, 'Parcial'] += 0.1
            print(f"Actualizado: {estudiante_aleatorio} ahora tiene {estudiantes.at[indice, 'Parcial']} en la columna 'Parcial'.")
        
        lista_estudiantes.remove(estudiante_aleatorio)
    else:
        print("Se han llamado a todos los estudiantes del grupo completo.")

# Guardar los cambios en el archivo CSV
def guardar_cambios():
    estudiantes.to_csv(path, index=False)
    print("Cambios guardados en el archivo CSV.")

# Ejemplo de uso
#while lista_estudiantes:
#    eval_oral()
#    guardar_cambios()


````

### **Pregunta 4: Demostración por Inducción Matemática (30 puntos)**



**Proposición**: Para todo $ n \in \mathbb{N} $, la suma de los primeros $ n $ números naturales es:

$$ 1 + 2 + 3 + \dots + n = \frac{n(n+1)}{2} $$

1. Interprete en lenguaje natural la proposición cuantificada dada:   
   La suma de los primeros $n$ números naturales es igual al semiproducto del ultimo sumando y su consecutivo. 

   $$ \text{suma de los primeros naturales} = \text{semiproducto del último sumando y su consecutivo}  $$ 

   

2. Demuestra la siguiente proposición utilizando el método de inducción matemática:
   

a) **Base de la inducción**: Demuestra que la proposición es verdadera para $ n = 1 $.

demostrar que la proposición es verdadera para $n = 1$, significa sustituir $n$ a ambos lados de la ecuación para ver si se convierte en una identidad o igualdad verdadera.  En este caso, la ecuación se convierte en: 

$$ 1 = \frac{1(2)}{2} $$

$$ 1 = 1 $$



b) **Hipótesis de inducción**: Supón que la proposición es verdadera para algún $ n = k $, es decir, que se cumple que:

$$ 1 + 2 + 3 + \dots + k = \frac{k(k+1)}{2} $$

c) **Paso inductivo**: Usa la hipótesis de inducción para demostrar que la proposición es verdadera para $ n = k + 1 $. Es decir, demuestra que:

$$ 1 + 2 + 3 + \dots + (k+1) = \frac{(k+1)((k+1)+1)}{2} $$

---

### Tabla para la demostración por inducción matemática

### Explicación de cada sección:
1. **Paso base**: El estudiante demuestra la validez de la proposición para el valor inicial, generalmente \( n = 1 \).
2. **Hipótesis inductiva**: El estudiante debe asumir que la proposición es verdadera para algún número \( n = k \). Esto se presenta en lenguaje formal.
3. **Tesis inductiva**: Se enuncia lo que el estudiante debe demostrar para \( n = k + 1 \).
4. **Demostración del paso inductivo**: Es donde el estudiante justifica que, partiendo de la hipótesis inductiva, puede deducir que la proposición es verdadera para \( n = k+1 \). Es el espacio más grande porque requiere mostrar el desarrollo paso a paso.
5. **Conclusión general**: El estudiante generaliza que la proposición es verdadera para todos los números naturales a partir de la validez del paso base y del paso inductivo.

Este formato proporciona una guía clara y estructurada para que los estudiantes no se pierdan en el proceso de la demostración por inducción matemática. Puedes adaptarlo a tus necesidades o agregar ejemplos según la proposición que se esté demostrando.

### Aplicaciones en Ciencia de Datos y Machine Learning

1. **Suma (conmutatividad y asociatividad)**:
   - En la ciencia de datos, la **suma** es esencial cuando agregamos características de conjuntos de datos o calculamos estadísticas descriptivas como la media, la suma acumulada y el total de ciertos valores. Las propiedades de **conmutatividad** y **asociatividad** garantizan que el orden en que sumamos no afecta el resultado, lo que es importante cuando se manipulan grandes volúmenes de datos.
   - En *machine learning*, la suma aparece en algoritmos de agregación como la regresión lineal, donde la suma ponderada de los valores de entrada determina la predicción.

   **Ejemplo en Python:**



   ```python
   import numpy as np

   data = np.array([3, 7, 2, 5])

   # Suma acumulada (propiedad conmutativa)
   total_sum = np.sum(data)
   print("Suma total:", total_sum)

   # Suma de todas las filas de una matriz (agregación)
   matrix = np.array([[1, 2], [3, 4], [5, 6]])
   row_sum = np.sum(matrix, axis=1)
   print("Suma por filas:", row_sum)
   ```

2. **Resta (no conmutativa)**:
   - La **resta** es utilizada en el cálculo de **errores** o **residuos** en muchos modelos de *machine learning*. Por ejemplo, en la regresión lineal se calculan los errores restando los valores predichos de los valores reales. La falta de conmutatividad en la resta garantiza que los signos de los errores reflejen adecuadamente las desviaciones de los datos predichos con respecto a los reales.

   **Ejemplo en Python:**

   ```python
   # Predicción vs valores reales (modelo lineal)
   valores_reales = np.array([10, 20, 30, 40])
   predicciones = np.array([12, 18, 29, 41])

   # Error de predicción (resta)
   errores = valores_reales - predicciones
   print("Errores:", errores)

   # Suma de los errores
   print("Suma de los errores:", np.sum(errores))
   ```

   - La **multiplicación** es fundamental en muchas áreas de ciencia de datos, especialmente en la **normalización** y **escalado** de características.    - En técnicas como el **gradiente descendente**, la multiplicación escalar de los gradientes por una tasa de aprendizaje optimiza los parámetros del modelo.

````python
# algoritmo para llamar a evaluación oral individual
import pandas as pd
import numpy as np

# Cargar la lista de estudiantes desde la URL
path = '/home/marco/Documentos/docencia/groups_list/calculo_vectorial.csv'
estudiantes = pd.read_csv(path)
estudiantes['Parcial'] = estudiantes['Parcial'].astype(float) 

# Convertir la columna 'Nombre' a una lista de Python
lista_estudiantes = estudiantes['Nombre'].tolist()

# Función para llamar a un estudiante aleatoriamente y registrar su nota de seguimiento
def eval_oral():
    if lista_estudiantes:
        estudiante_aleatorio = np.random.choice(lista_estudiantes)
        print(f"Llamando a {estudiante_aleatorio}...")
        asistencia = input(f"¿Se expresó bien y correco {estudiante_aleatorio}? (s/n): ").strip().lower()
        
        if asistencia == 's':
            # Obtener el índice del estudiante en el DataFrame
            indice = estudiantes[estudiantes['Nombre'] == estudiante_aleatorio].index[0]
            # Incrementar su valor en la columna 'Parcial'
            estudiantes.at[indice, 'Parcial'] += 0.1
            print(f"Actualizado: {estudiante_aleatorio} ahora tiene {estudiantes.at[indice, 'Parcial']} en la columna 'Parcial'.")
        
        lista_estudiantes.remove(estudiante_aleatorio)
    else:
        print("Se han llamado a todos los estudiantes del grupo completo.")

# Guardar los cambios en el archivo CSV
def guardar_cambios():
    estudiantes.to_csv(path, index=False)
    print("Cambios guardados en el archivo CSV.")

# Ejemplo de uso
while lista_estudiantes:
    eval_oral()
    guardar_cambios()


````

````python
import csv

def recolectar_datos_autoavaluacion():
    # Definir el nombre del archivo CSV
    nombre_archivo = 'autoevaluacion_estudiantes.csv'
    
    # Definir los encabezados para el archivo CSV
    encabezados = ['Nombre', 'Fortalezas', 'Áreas de Mejora']
    
    # Solicitar los datos del estudiante
    nombre = input("Ingrese su nombre: ")
    fortalezas = input("Describa sus fortalezas: ")
    areas_mejora = input("Describa las áreas en las que necesita mejorar: ")
    
    # Crear una lista con los datos del estudiante
    datos = [nombre, fortalezas, areas_mejora]
    
    # Abrir el archivo CSV para agregar los datos
    with open(nombre_archivo, mode='a', newline='') as archivo_csv:
        escritor_csv = csv.writer(archivo_csv)
        
        # Verificar si el archivo está vacío para escribir los encabezados
        archivo_csv.seek(0, 2)  # Mover el cursor al final del archivo
        if archivo_csv.tell() == 0:  # Verificar si el archivo está vacío
            escritor_csv.writerow(encabezados)
        
        # Escribir los datos en el archivo CSV
        escritor_csv.writerow(datos)
    
    print("La autoevaluación se ha guardado correctamente.")

# Ejecutar la función para recolectar datos
recolectar_datos_autoavaluacion()
````
